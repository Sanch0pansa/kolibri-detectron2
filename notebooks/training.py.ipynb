## Training on single class

```python
sd = Dataset("../data/updated_dataset/dataset.yaml", single_class=True)
sd.register("single-class2")
```

```python
Faster_RCNN_R50_1x = RCNN(
    "COCO-Detection/faster_rcnn_R_50_FPN_1x.yaml",
    output_dir="faster_rcnn_R_50_1x-output",
)
```

```python
Faster_RCNN_R50_1x.train(
    sd.train_dataset_name,
    max_iterations=280
)
```

```python
Faster_RCNN_R50_1x.config_model()
Faster_RCNN_R50_1x.init_predictor()
Faster_RCNN_R50_1x.visualize_prediction(d.valid_dataset[1]['file_name'], d.get_metadata("val"))
```

```python
RetinaNet_R50_FPN_1x = RetinaNet(
    "COCO-Detection/retinanet_R_50_FPN_1x.yaml",
    output_dir="retinanet_R_50_FPN_1x_single-class-output",
)
```

```python
RetinaNet_R50_FPN_1x.train(
    sd.train_dataset_name,
    max_iterations=1000
)
```

```python
RetinaNet_R50_FPN_1x.cfg.MODEL.RETINANET.SCORE_THRESH_TEST = 0.5
```

```python
RetinaNet_R50_FPN_1x.config_model()
RetinaNet_R50_FPN_1x.init_predictor()
RetinaNet_R50_FPN_1x.visualize_prediction(d.valid_dataset[1]['file_name'], d.get_metadata("val"))
```

# Test with bigger threshold

```python
sd = Dataset("../data/updated_dataset/dataset.yaml", single_class=True)
sd.register("single-class")
```

```python
RetinaNet_R50_FPN_3x = RetinaNet(
    "COCO-Detection/retinanet_R_50_FPN_3x.yaml",
    output_dir="retinanet_R_50_FPN_3x_single_class_big_threshold-output",
)
```

```python
RetinaNet_R50_FPN_3x.train(
    sd.train_dataset_name,
    max_iterations=280,
    score_thresh_test=0.5,
)
```
